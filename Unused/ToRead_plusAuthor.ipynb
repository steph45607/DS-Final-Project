```python
def clean_publish_year(raw_year):
    if type(raw_year) is None:
        return "Missing"
    result = re.search(r"(\b\d+)$", raw_year)
    if result == None:
        return 0
    else:
        return result.group(1)

def clean_rating(raw_rating):
    if type(raw_rating) is None:
        return 0
    result = re.search(r"(\b\d+.\b\d+)", raw_rating)
    if result == None:
        return 0
    else:
        return result.group(1)

def clean_raters(raw_raters):
    if type(raw_raters) is None:
        return 0
    result = re.search(r"((?:\d+,?\d+)*) ratings", raw_raters)
    if result == None:
        return 0
    else:
        return result.group(1)
    
def clean_genres(raw_genres):
    genreList = []
    if raw_genres == None:
        return "Missing"
    else:
        for genre in raw_genres:
            if  genre == None:
                continue
            else:
                genreList.append(genre.find("a").get_text())
        return genreList

def clean_works(raw_works):
    if type(raw_works) is None:
        return 0
    result = re.search(r"((\d+(\,)?)+)", raw_works)
    if result == None:
        return 0
    else:
        return result.group(1)
        
def clean_year(raw_year):
    if type(raw_year) is None:
        return 0
    result = re.search(r"Published\s*[A-z]*\s[0-9]*([a-z]{2}).([0-9]{4})", raw_year)
    if result == None:
        return 0
    else:
        return result.group(2)
# def clean_readers(raw_readers):
#     result = re.search(r"(\b\d+)", raw_readers)
#     if result == None:
#         return "None"
#     else:
#         return result.group(1)

def solve_nonetype_error(value):
    if value is not None:
        return value.text.strip()
    else:
        return "Missing"

```

```python
books = []
for page in range(21, 26):
    link = "https://www.goodreads.com/shelf/show/to-read?page=" + str(page)
    response = requests.get(link, cookies = cookies, headers=headers)
    html_soup = BeautifulSoup(response.text, 'html5lib')
    book_container = html_soup.find('div', class_ = "leftContainer")
    book_elements = book_container.find_all('div', class_ = "elementList")
    print("this is page " , page)
    for book in book_elements:
        bookTitle = solve_nonetype_error(book.find('a', class_ = "bookTitle"))
        # print(bookTitle)
        bookAuthor =solve_nonetype_error(book.find('a', class_ = "authorName"))
        # print(bookAuthor)

        author_link = book.find('a', class_ = "authorName")['href']
        # print(author_link)
        getAuthor = requests.get(author_link, cookies=cookies2, headers=headers2)
        html_author = BeautifulSoup(getAuthor.text, 'html5lib')

        avg_rating = solve_nonetype_error(html_author.find("span", class_ = "average"))
        # print(avg_rating)
        num_raters = solve_nonetype_error(html_author.find("span", class_ = "votes"))
        # print(num_raters)
        num_reviews = solve_nonetype_error(html_author.find("span", class_ = "count"))
        # print(num_reviews)
        authorTest = html_author.find("div", class_="hreview-aggregate")
        print(authorTest)
        if authorTest is None:
            author_works = "Missing"
        else:
            author_works = solve_nonetype_error(authorTest.find("a", href = True))
            author_works = clean_works(author_works)


        year = clean_publish_year(solve_nonetype_error(book.find('span', class_ = "greyText smallText")))
        # print(year)
        rating = clean_rating(solve_nonetype_error(book.find('span', class_ = "greyText smallText")))
        # print(rating)
        raters = str(clean_raters(solve_nonetype_error(book.find('span', class_ = "greyText smallText")))).replace(",", "")
        int(raters)
        # print(raters)

        link = book.a['href']
        completeLink = "https://www.goodreads.com" + link
        getGenre = requests.get(completeLink,  cookies = cookies3, headers=headers3)
        genreParsing = BeautifulSoup(getGenre.text, 'html5lib')
        genres = clean_genres(genreParsing.findAll("div", class_=  "elementList", style = None))
        genreStr = ",".join(genres)
        # print(genreStr)
        book_data = [bookTitle, bookAuthor,avg_rating, num_raters, num_reviews, author_works , year,  rating, raters, genreStr]
        books.append(book_data)
books = pd.DataFrame(books, columns=["Title", "Author", "Author Average Rating", "Author Votes", "Review Count", "Distinct Works", "Year", "Rating", "Raters", "Genres"])
books.head()
```
