```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
df = pd.read_csv("CombinedBooksRandomized.csv")
df.head()
```

```python
pearson = df.corr(method='pearson')
pearson
```

```python
sb.heatmap(pearson,
xticklabels=pearson.columns,
yticklabels=pearson.columns,
cmap='RdBu_r',
annot=True,
linewidths=0.5)
```
