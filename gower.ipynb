```python
import pandas as pd
import gower
from sklearn.cluster import DBSCAN
```

```python
df = pd.read_csv("CombinedBooksRandomized.csv")
df = df.loc[:,"Author Average Rating":"Genres"]
df
```

```python
distance_matrix = gower.gower_matrix(df)
distance_matrix
```

```python
dbscan_cluster = DBSCAN(eps=0.3, min_samples=10, metric="precomputed")

dbscan_cluster.fit(distance_matrix)
df["cluster"] = dbscan_cluster.labels_
```

```python
df
```

```python
df.to_csv("DBSCAN.csv", index = False)
```
