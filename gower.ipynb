```python
import pandas as pd
import gower
from sklearn.cluster import DBSCAN
```

```python
df = pd.read_csv("CombinedBooksClass.csv")
df = df.loc[:,"Author Average Rating":"Raters"]
df
```

```python
distance_matrix = gower.gower_matrix(df)
distance_matrix
```

```python
dbscan_cluster = DBSCAN(eps=0.3, min_samples=2, metric="precomputed")

dbscan_cluster.fit(distance_matrix)
df["cluster"] = dbscan_cluster.labels_
```

```python
df
```

```python
df.to_csv("DBSCAN.csv", index = False)
```
